```r
# For sex & address & failures & freetime I will use Mode function
get_mode <- function(v) {
  uniq_vals <- na.omit(unique(v))
  uniq_vals[which.max(tabulate(match(v, uniq_vals)))]
}
```

```r
mode_sex <- get_mode(DF$sex)

DF$sex[is.na(DF$sex)] <- mode_sex
```

```r
mode_address <- get_mode(DF$address)

DF$address[is.na(DF$address)] <- mode_address
```

```r
mode_failures <- get_mode(DF$failures)

DF$failures[is.na(DF$failures)] <- mode_failures
```

```r
mode_freetime <- get_mode(DF$freetime)

DF$freetime[is.na(DF$freetime)] <- mode_freetime
```

# 5.Exploratory Data Analysis (EDA)

* Identify & Remove Outliers

```r
boxplot(DF$absences,
        main = "Absence for each student",
        ylab = "Count",
        col = "lightblue",
        border = "darkblue")
```

```r
# Calculate Q1 and Q3
Q1 <- quantile(DF$absences, 0.25, na.rm = TRUE)
Q3 <- quantile(DF$absences, 0.75, na.rm = TRUE)

# Calculate IQR
IQR <- Q3 - Q1

# Define lower and upper bounds
lower_bound <- Q1 - 1.5 * IQR
upper_bound <- Q3 + 1.5 * IQR

# Filter the data to remove outliers
DF <- DF[DF$absences >= lower_bound & DF$absences <= upper_bound, ]
```

```r
boxplot(DF$absences,
        main = "Boxplot of Absences",
        ylab = "Number of Absences",
        col = "lightblue",
        border = "darkblue")
```

### Looking for Correlations

```r
install.packages("corrplot", repos = "https://cloud.r-project.org")
```

```r
library(dplyr)
library(corrplot)

numeric_df <- DF %>% select(where(is.numeric))


cor_matrix <- cor(numeric_df, use = "complete.obs")


corrplot(cor_matrix,
         method = "color",
         type = "upper",
         tl.cex = 0.6,
         number.cex = 0.5,
         addCoef.col = "black",
         col = colorRampPalette(c("blue", "white", "red"))(200),
         mar = c(0, 0, 2, 0))
```

Key Insights :
1. Strong Relationship between G2 and G3
There is a very strong positive correlation between the second period grade (G2) and the final grade (G3).
* This indicates that students who perform well earlier are very likely to maintain their performance.

2. Previous Performance Matters (G1 & G2)
Both G1 and G2 show strong positive relationships with G3, suggesting that academic performance is consistent over time.

3. Negative Impact of Failures
A moderate negative relationship exists between failures and G3.
* Students with more past failures tend to achieve lower final grades.

4. Weak Effect of Study Time
Study time shows only a weak positive relationship with G3, indicating that simply increasing study hours does not strongly guarantee better performance.

5. Minimal Impact of Social Life (goout)
The level of going out with friends has a very weak negative relationship with G3, suggesting limited impact on academic performance.

6. Absences Have Limited Effect
Absences show very weak correlation with G3, meaning attendance alone is not a strong predictor of performance in this dataset.
